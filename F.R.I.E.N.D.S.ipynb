# 2. Data Cleansing & Data Manipulation

```python
episode = episode.drop(235)
```

```python
episode.info()
```

```python
imdb.info()
```

```python
episode = episode.convert_dtypes()
episode.info()
```

```python
episode['original_air_date'] = pd.to_datetime(episode['original_air_date'])
episode.info()
```

```python
imdb = imdb.convert_dtypes()
imdb.info()
```

```python
imdb['original_air_date'] = pd.to_datetime(imdb['original_air_date'])
imdb.info()
```

There are same 4 columns between episode and imdb dataframe (`season`, `episode_num`, `title`, and `original_air_date`). It will be more simple if we drop those columns.

```python
# Remove 4 columns
imdb = imdb.drop(columns=['season','episode_num','title','original_air_date'])
imdb.head()
```

```python
# Combine episode and imdb dataframes become one dataframe
friends = pd.concat([episode, imdb], axis=1)
friends.head()
```

```python
friends.info()
```

# 3. Export file into spreadsheet

```python
friends.to_excel('friends.xls') 
files.download('friends.xls')
```
